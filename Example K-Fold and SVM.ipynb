```python
dict_file_sample = {i:list() for i in list_file_test}
for i in list_file_test:
    location_folder = directoryTest+"/"+i
    temp_folder = os.listdir(location_folder)
    for j in temp_folder:
        print(location_folder+"/"+j)
        img_file = cv2.imread(location_folder+"/"+j)
        img_file = img_file/255
        namefile = j.split("_")
        dict_file_sample[namefile[0]].append(img_file)
```

```python
label = list(dict_file.keys())
```

```python
x = list()
y = list()
for i in label:
    for j in dict_file[i]:
        x.append(j)
        y.append(i)
```

```python
x_sample = list()
y_sample = list()
for i in label:
    for j in dict_file[i]:
        x_sample.append(j)
        y_sample.append(i)
```

```python
x_np = np.array(x)
y_np = np.array(y) 
```

```python
x_np_sample = np.array(x_sample)
y_np_sample = np.array(y_sample) 
```

```python
kf = KFold(n_splits=5, shuffle=True)
```

```python
y_np
```

```python
def train_multipleclassification(Model, x_train, y_train, x_test, y_test):
    mdl = Model
    print("fit start")
    self = mdl.fit(X=x_train, y=y_train)
    print("fit finish")
    result = mdl.predict(x_test)
    self_res = mdl.predict(x_train)
    
    score = mdl.score(x_test, y_test)
    self_score = mdl.score(x_train, y_train)
    print()
    print("train_score")
    print(score)
    print("test_score")
    print(self_score)
    return mdl
    
```

```python
from sklearn import svm
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.8)
ctr = 1
for train_index, test_index in kf.split(x):
    print("TRAIN:", train_index, "TEST:", test_index)
    print("fold : ", ctr)
    ctr += 1
    print("Start PCA")
    
    flatten = [i.flatten() for i in x_np_sample]
    pca = pca.fit(flatten)
    flatten_train = [i.flatten() for i in np.take(x_np, train_index)]
    pca_result = pca.transform(flatten_train)
    flatten_test = [i.flatten() for i in np.take(x_np, test_index)]
    pca_result_test = pca.transform(flatten_test)
    print("PCA finish")
    print("")
    mdl = train_multipleclassification(svm.SVC(gamma='scale', decision_function_shape='ovo'),
                                       pca_result, np.take(y_np, train_index), pca_result_test, np.take(y_np, test_index))

    
    
    
    
```
