```python
hog_features=[]
hog_images=[]

for i in df['file'] :
#     print(i)
    img = cv2.imread(i)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    fd, hog_image = hog(img, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True)
    hog_images.append(hog_image)
    hog_features.append(fd)
    
```

```python
labels =  np.array(df['label']).reshape(len(df['label']),1)
```

```python
hog_feature = np.array(hog_features)
```

```python
data_frame = np.hstack((hog_feature,labels))
```

```python
hog_feature.shape
```

```python

from sklearn.model_selection import LeaveOneOut,KFold
```

```python
kf = KFold(n_splits=10, shuffle=True)
```

```python
ctr = 1
for train_index, test_index in kf.split(hog_feature):

    x_train = hog_feature[train_index]
    x_test = hog_feature[test_index]
    y_train = labels[train_index]
    y_test = labels[test_index]

    clf = svm.SVC(kernel='linear', C=10).fit(x_train, y_train)
    print("Kfold : "+str(ctr))
    ctr += 1
    print(clf.score(x_test, y_test))
    
    
    
```
